### 問1.1 FizzBuzz
<font size="3" >[アルゴリズム]</font><br>
数字を1からnまでを順に出力する。ただし、3で割り切れる時は、**"Fizz"**、5で割り切れる時は**"Buzz"**、15で割り切れる時は、**"FIzzBuzz"**と数字の代わりに出力する。以下はFizzBuzzゲームの進行例である。


>1 2, Fizz 4 Buzz Fizz 7 8 FizBuzz 11 Fizz 13 14 FizzBuzz 16 17 Fizz 19 Buzz ....

### (1) 1から100までの値でFizzBuzzゲームをするプログラムを実装せよ。
 コメント
 pythonでは、ifの条件に()は書かなくてもいいです。
 

````python
##FizzBuzz 1~100の数字でFIzzBuzzゲームを実装する(3/3点)
for i in range(100):
    if (i + 1) % 15 == 0: 
        print("FizzBuzz")
    elif (i + 1) % 5 == 0:
        print("Buzz")
    elif (i + 1) % 3 == 0:
        print("Fizz")
    else :
        print(i + 1)
````

### (2) 引数として変数nを受け取ると、1~nまでのFizzBuzzゲームの結果を出力する関数FizzBuzzを実装せよ。

````python
## def FizzBuzz(n)を実装する。(2/２点)
def FizzBuzz(n):
    for i in range(n):
        if (i + 1) % 15 == 0: 
            print("FizzBuzz")
        elif (i + 1) % 5 == 0:
            print("Buzz")
        elif (i + 1) % 3 == 0:
            print("Fizz")
        else :
            print(i + 1)



## 実行結果の確認
FizzBuzz(100)
````

>function 深さ優先探索(v)  
　　S ← 空のスタック

## 問1.4 クラス
名前と血液型と体重と身長をインスタンス変数とし、BMIを計算をするcalc_bmiメソッドと、指標に基づいて体型を評価するeval_bmiメソッドと、情報を表示するshowメソッドをもつクラス「Person」を作成したい。BMIは体重 $w[kg]$, 身長$h[m]$とすると、$$BMI = w/h^2$$と表すことができる。また、$$18.5 \leq BMI < 25.0$$を標準BMIとし、標準BMI未満は痩せ型、標準BMI以上は肥満型とする。



ここでは、例えば、名前が○○, 血液型が?型, 体重45.0kg, 身長1.50mの人は
```
○○さんの血液型は?型で、BMIは20.00で標準体型です。
```
と表示されるようにしたい。  

### 以下のコードを埋めよ。

コメント
　関数の出力値はreturnを使って戻り値にすると見やすくなります。

````python
class Person:
    def __init__(self, name, blood_type, weight, height):
        # ここに入力してください。（1/1点）
        self.name = name
        self.blood_type = blood_type
        self.weight = weight
        self.height = height
    def calc_bmi(self):
        # ここに入力してください。（1/1点）
        return self.weight / (self.height * self.height)
    def eval_bmi(self):
        # ここに入力してください。（1/1点)
        BMI = self.calc_bmi()
        if BMI >= 25.0:
            return "肥満型"
        elif BMI <= 18.5:
            return "ヤセ型"
        else :
            return "標準型"
    def show(self):
        # ここに入力してください。（1/1点）
        print(self.name, "さんの血液型は", self.blood_type, "型")
        print("BMIは%.2fで%sだよ" % (self.calc_bmi(), self.eval_bmi()))
# 名前が「森」、血液型は「B型」、体重は「58.0kg」、 身長は「1.75m」のインスタンスを生成して、
# 上の例のように出力する(BMIは小数点以下2桁まで表示)コードを書いてください。（1/1点）

person = Person("森", "B", 58.0, 1.75)
person.show()
````

このファイルを解析する以下の問いに答えよ

````python
#一度だけ実行すること
import random
random.seed(1)
with open('fruits.txt','w') as f:
    for _ in range(1000):
        f.write(random.choice(['apple', 'orange', 'strawberry', 'cherry', 'plum','banana', 'grape','melon', 'peach', 'watermelon'])+'\n')
````

### (1)`fruits.txt`からデータを読み込み、list型の変数`fruits`を作成せよ
>fruits = [apple,banana, ....]

<font size="3" >[ヒント]</font>  
ファイルの読み込みには`with open('filepath','r') as f`を使う

````python
## listにする(1/1点)
fruits = [
    
]
with open('./fruits.txt','r') as f:
    _fruits = f.readlines()
    for fruit in _fruits:
        fruits.append(fruit.replace('\n', ''))
    f.close()
fruits
````

### (2)`fruits`にある果物の種類を表示するlist型の変数`fruits_list`の作成及び、果物の数を表示せよ。

````python
## fruits_listの作成(2/2点)
fruits_list = []
for fruit in fruits:
    if fruit not in fruits_list:
        fruits_list.append(fruit)
print(fruits_list)

##フルーツの種類の出力(1/1点)
print(len(fruits_list))
````

### (3)`fruits`に果物がそれぞれ何個ずつあるのかを表示するdict型の変数`fruits_num`を作成せよ
> fruits_num = {'peach': 10, 'melon'30, ....}

 コメント
 はじめのfruits_numの作成では、何のフルーツがあるのかも分かってない状態で変数を作ってください。はじめからdict型で定義する方法でも、他の型で定義した後dict型に変換する方法でも、どちらでもできます。
 それぞれのフルーツの数は正解です。

````python
## fruits_numの作成
fruits_num = dict.fromkeys(fruits_list,0)

for fruit in fruits:
    fruits_num[fruit] += 1

fruits_num
````
